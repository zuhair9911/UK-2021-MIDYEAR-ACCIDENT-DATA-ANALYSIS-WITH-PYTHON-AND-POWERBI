```python
import pandas as pd
import matplotlib.pyplot as plt

#reading excel file

accident_data = pd.read_excel("UK-Accidents.xlsx")
accident_data


```

```python
#dropping columns
accident_data1 = accident_data.drop(columns = ['status','accident_year','trunk_road_flag','urban_or_rural_area','lsoa_of_accident_location'])
accident_data1
```

```python
#Total number of excidents
total_accidents = accident_data1['accident_index'].count()
print("Total number of accidents: " + str(total_accidents))

#Total number of casualties
casualties = accident_data1['number_of_casualties'].sum()
print("Total number of causalties: " + str(casualties))

#Total number of vehicles involved
vehicles_involved = accident_data1['number_of_vehicles'].sum()
print("Total number of vehicles involved: " + str(vehicles_involved))
```

```python
#count of accidents by month
accident_data1.groupby(['Month'])['Month'].count()
```

```python
#count of accidents by day of the week
accident_data1.groupby(['day_of_week'])['day_of_week'].count()
```

```python
#count of accidents by light conditions
accident_data1.groupby(['light_conditions'])['light_conditions'].count()
```

```python
#count of accidents by severity
accident_data1.groupby(['accident_severity'])['accident_severity'].count()
```

```python
#count of accidents by road surface conditions
accident_data1.groupby(['road_surface_conditions'])['road_surface_conditions'].count()
```

```python
#count of accidents by time interval
accident_data1.groupby(['Time_Group'])['Time_Group'].count()
```
